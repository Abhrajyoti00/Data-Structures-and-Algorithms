```python
class Graph:
    def __init__(self, nVertices):
        self.nVertices = nVertices
        self.adjMatrix = [[0 for j in range(self.nVertices)] for j in range(self.nVertices)]
    
    def addEdge(self, v1,v2):
    
        self.adjMatrix[v1][v2] = 1
        self.adjMatrix[v2][v1] = 1
    
    
    def __hasPathHelper(self, sv, v2, visited):
        if (sv == v2) or self.adjMatrix[sv][v2] > 0:
            return True
        visited[sv] = True

        for i in range(self.nVertices):
            if ((self.adjMatrix[sv][i] > 0) and (visited[i] is False)):  # Not visited and edge is present:
                if self.__hasPathHelper(i, v2,visited):
                    return True

        return False

    def hasPath(self, v1,v2,v):
        visited = [False for i in range(v)]
        return self.__hasPathHelper(v1,v2,visited)
        

    def __getPathHelper(self, sv, v2, solution, visited):

        visited[sv] = True

        if (sv == v2) or self.adjMatrix[sv][v2]>0:
           solution.append(v2)
           solution.append(sv)
           return solution

        solution.append(v2)
        solution.append(sv)

        for i in range(self.nVertices):
            if ((self.adjMatrix[sv][i]>0) and (visited[i] is False)):
                if self.__getPathHelper(i, v2, solution, visited):
                	solution.append(sv)
                	return solution
        return False
                            

    def getPath(self, sv, v2, solution):
        visited = [False for i in range(self.nVertices)]
        solution = []
        return self.__getPathHelper(sv, v2, solution, visited)
        
        
ve = list(map(int,input().split()))
v = ve[0]
e = ve[1]
g = Graph(v+1)
for i in range(e):
    v1v2 = list(map(int,input().split()))
    v1 = v1v2[0]
    v2 = v1v2[1]
    g.addEdge(v1,v2)



v1v2 = list(map(int,input().split()))
v1 = v1v2[0]
v2 = v1v2[1]

solution = []
path = g.getPath(v1,v2,solution)
if path:
    for i in g.getPath(v1,v2,solution):
        print(i, end = ' ')
```
