```python
import queue

class Graph:
    def __init__(self, nVertices):
        self.nVertices = nVertices
        self.adjMatrix = [[0 for j in range(nVertices)] for i in range(nVertices)]

    def addEdge(self, v1,v2):
    
        self.adjMatrix[v1][v2] = 1
        self.adjMatrix[v2][v1] = 1
    
    def removeEdge(self, v1, v2):
        if self.containsEdge(v1,v2) is False:
            return
        self.adjMatrix[v1][v2] = 0
        self.adjMatrix[v2][v1] = 0

    def containsEdge(self, v1, v2):
        return True if self.adjMatrix[v1][v2]>0 and self.adjMatrix[v2][v1] > 0 else False

    def __str__(self):
        return str(self.adjMatrix)


    def __dfsHelper(self, sv, visited):
        print(sv)
        visited[sv] = True

        for i in range(self.nVertices):
            if self.adjMatrix[sv][i] > 0 and visited[i] is False:  # Not visited and edge is present:
                self.__dfsHelper(i, visited)

    def dfs(self):
        visited = [False for i in range(self.nVertices)]
        self.__dfsHelper(0,visited)


    def __bfs (self, sv, visited):
        q = queue.Queue()
        q.put(sv)
        visited[sv] = True
        while q.empty() is False :
            u = q.get()
            print(u, end= " " )
            for i in range(self.nVertices) :
                if self.adjMatrix[u][i] > 0 and visited[i] is False :
                    q.put(i)
                    visited[i] = True
    def bfs (self):
        visited = [ False for i in range(self.nVertices)]
        for i in range(self.nVertices):
            if visited[i] is False :
                self.__bfs(i, visited)


    def hasPath(self, v1, v2):
        visited = [False for i in range(self.nVertices)]
        if self.containsEdge(v1,v2):
            return True
        else:
            for i in range(self.nVertices):
                if visited[i] == False:
                    if self.hasPath(i, v2):
                        visited[i] = True
                        return True
                        
        return False

ve = list(map(int,input().split()))
v = ve[0]
e = ve[1]
g = Graph(v)
for i in range(e):
    v1v2 = list(map(int,input().split()))
    v1 = v1v2[0]
    v2 = v1v2[1]
    g.addEdge(v1,v2)

v1v2 = list(map(int,input().split()))
v1 = v1v2[0]
v2 = v1v2[1]
if g.hasPath(v1, v2):
    print('true')
else:
    print('false')
```
