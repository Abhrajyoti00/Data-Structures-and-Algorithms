```python
def fib(n):
    dp[0] = 0
    dp[1] = 1

    for i in range(2,n+1):
        dp[i] = dp[i-1] + dp[i-2]
    return dp[n]

n = int(input())
dp = [0 for i in range (n+1)]
print(fib(n))
```
